```python
import cv2, numpy as np
from matplotlib import pyplot as plt
from mtcnn import MTCNN
```

```python
detector = MTCNN()
```

```python
input_dir = '/content/drive/My Drive/Virat/validation'
output_dir = '/content/drive/My Drive/Virat/cropped_validation'
```

```python
os.makedirs(output_dir, exist_ok=True)
```

```python
desired_image = (128, 128)
```

```python
def img_processing(img_path, output_path):
  img = cv2.imread(img_path)
  image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  detections = detector.detect_faces(image_rgb)
  if detections:
    selected_face = detections[0]
    x, y, width, height = selected_face['box']
    cropped_face = image_rgb[y:y+height, x:x+width]
    resized_face = cv2.resize(cropped_face, desired_image)
    cv2.imwrite(output_path, resized_face)
    print(f"Image processed and saved: {output_path}")
  else:
    print(f"No face detected in {img_path}")
```

```python
for filename in os.listdir(input_dir):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename)
        img_processing(input_path, output_path)
```
